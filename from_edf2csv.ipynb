```python
import numpy as np
import mne
import pandas as pd
import matplotlib.pyplot as plt
from os import listdir
from os.path import isfile, join
```

```python
def edf2csv(mypath):
    #SubjectXX_1.edf - EEG before mental arithmetic task
    #SubjectXX_2.edf - EEG during the mental arithmetic task
    onlyfiles = [join(mypath, f) for f in listdir(mypath) if isfile(join(mypath, f)) and f.endswith('.edf')]
    filesName = [f for f in listdir(mypath) if isfile(join(mypath, f)) and f.endswith('.edf')]

    for ii,jj in enumerate(onlyfiles):
        
        data = mne.io.read_raw_edf(onlyfiles[ii])
        raw_data = data.get_data()
        # you can get the metadata included in the file and a list of all channels:
        info = data.info
        channels = data.ch_names
        if filesName[ii].split('.')[0][-1] == '1':
            temp = np.zeros([1,raw_data.shape[1]],dtype='int')

        elif filesName[ii].split('.')[0][-1] == '2':
            temp = np.ones([1,raw_data.shape[1]],dtype='int')

        df = pd.DataFrame(raw_data.T,columns=channels)
        df['Action'] = temp[0]
        df.to_csv('eeg_csv\\'+filesName[ii].split('.')[0]+'.csv', index=False)
#     return df
```
